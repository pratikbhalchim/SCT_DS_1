```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Country Name": [
        "India", "United States", "China", "Japan", "Germany",
        "United Kingdom", "France", "Canada", "Australia", "Brazil"
    ],
    "Country Code": [
        "IND", "USA", "CHN", "JPN", "DEU",
        "GBR", "FRA", "CAN", "AUS", "BRA"
    ],
    "Year": [1999]*10,
    "Population": [
        1018737000, 279040000, 1252735000, 126631000, 82100243,
        58892514, 60912500, 30685730, 18926000, 172006000
    ]
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv("country_population_1999.csv", index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(
    df["Country Name"],
    df["Population"],
    color="red"
)
plt.xticks(rotation=45, ha="right")
plt.title("Country-wise Population Distribution (1999)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.hist(
    df["Population"],
    bins=5,
    color="pink",
    edgecolor="black"
)
plt.title("Population Distribution Histogram (1999)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
```
